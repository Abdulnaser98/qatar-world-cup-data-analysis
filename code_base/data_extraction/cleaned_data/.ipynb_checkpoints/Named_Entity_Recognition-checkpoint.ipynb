# ***Named Entity Recognition***

The goal of named entity recognition is to find all entities that are interessant , like personen , organisaitons 
and other important entites in order to conduct further analysis on the data ,like finding how are the general 
sentiment of the media outlets about the obsovered entites 

```python
# import libraries 
import pandas as pd 
import nlp
import spacy 
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
# read the three cleaned data sources 
aljazerra_cleaned_data = pd.read_csv("aljazerra_cleaned_data.csv")
bbc_cleaned_data = pd.read_csv("bbc_cleaned_data.csv")
guardian_cleaned_data = pd.read_csv("guardian_cleaned_data.csv")
```

### Extract Named entity recognition entities 

```python
def extract_NER_entities(data):
    nlp = spacy.load('en_core_web_sm')
    named_entities = []
    for content in tqdm(data["main_content"]):
        temp_entity_name = ''
        temp_named_entity = None
        content = nlp(content)
        for word in content:
            term = word.text 
            tag = word.ent_type_
            if tag:
                temp_entity_name = ' '.join([temp_entity_name, term]).strip()
                temp_named_entity = (temp_entity_name, tag)
            else:
                if temp_named_entity:
                    named_entities.append(temp_named_entity)
                    temp_entity_name = ''
                    temp_named_entity = None

    entity_frame = pd.DataFrame(named_entities, 
                            columns=['Entity Name', 'Entity Type'])
    print("the entity frame has the shape {}".format(entity_frame.shape))
    return entity_frame 
```
