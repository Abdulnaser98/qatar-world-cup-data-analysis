# Sentiment analysis

The data is already cleaned , so we do not need to clean it , we can use it directly 

```python
sia = SIA()
results = []
for line, url in zip(guardian_data['main_content'],guardian_data['url']) :
    pol_score = sia.polarity_scores(line)
    pol_score['headline'] = line
    pol_score['url'] = url
    results.append(pol_score)
```

```python
# Creating a new dataframe of only the polarity score, the headline and the source of the news 
headlines_polarity = pd.DataFrame.from_records(results)
# word count of news headlines is calculated
headlines_polarity['word_count'] = headlines_polarity['headline'].apply(lambda x: len(str(x).split()))
```

```python
sorted_headlines_polatity = headlines_polarity.sort_values(by=['pos'], ascending=False)
```

```python
sorted_headlines_polatity.head(40)
```

```python
guardian_data['title'] = guardian_data['title'].fillna('')
negatives = []
positives =[]
neutral =[]

for line,title in zip(guardian_data['main_content'],guardian_data['title']):
    title_score = sia.polarity_scores(title)
    main_content_score = sia.polarity_scores(line)
    negatives.append(((title_score['neg'])+(main_content_score['neg']))/2)
    positives.append(((title_score['pos'])+(main_content_score['pos']))/2)
    neutral.append(((title_score['neu'])+(main_content_score['neu']))/2)
guardian_data['negative'] = negatives 
guardian_data['positive'] = positives 
guardian_data['neutral'] = neutral 
    
```

```python
print("negative mean is : ")
print(guardian_data['negative'].mean())
print("positive mean is : ")
print(guardian_data['positive'].mean())
print("neutral mean is : ")
print(guardian_data['neutral'].mean())
```
