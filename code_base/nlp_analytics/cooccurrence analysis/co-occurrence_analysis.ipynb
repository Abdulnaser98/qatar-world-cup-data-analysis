# ***Import libraries***:

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

# ***co_occurrence Analysis:***

```python
def generate_coocurrance_analysis_plot(data,negative_words,positive_words, media_outlet_name,word_to_obsorve):
    
    fig, ax = plt.subplots(figsize=(10, 9))

    colors = ['green' if x in positive_words else 'red' if x in negative_words else 'gray' for x in data['names']]

    # Plot the DataFrame
    plt.barh(data['names'], data['values'], color=colors )

    # Set the axis labels and title
    ax.set_xlabel('Frequency')
    ax.set_ylabel('Label')
    ax.set_title(media_outlet_name +' co occurrence Analysis of the word ' + word_to_obsorve)

    plt.show()
```

##### Comparison of the cooccurrence results between the three media outlets regarding the subtopic "worker rights"

##### import data 

```python
path_to_evaluation_folder = "../../evaluation"
guardian_worker_rights_data = pd.read_csv(path_to_evaluation_folder + "/guardian/cooccurance Analysis/guardian_worker_rights.csv")
bbc_worker_rights_data = pd.read_csv(path_to_evaluation_folder + "/bbc/cooccurance Analysis/bbc_worker_rights.csv" )
aljazerra_worker_rights_data = pd.read_csv(path_to_evaluation_folder + "/aljazerra/cooccurance Analysis/aljazerra_worker_rights.csv")
```

Specify the top coocurring negative and positive words that come with the word "worker" 
in the context of the subtopic of "worker rights".

```python
guardian_neg_words = ["abuse","corruption","discriminatory","treatment","die","guardian","kafala","amnesty",
                       "death","exploit","exploitation"]
guardian_pos_words = ["protect","change","reform","improve"]

bbc_neg_words = ["wage","compensation","amnesty","death","abuse"]
bbc_pos_words= ["reform","ensure","change"]

aljazerra_pos_words = ["change","improve","ensure","protect","improve","protection"]
aljazerra_neg_words = ["abuse","complaint","death","salary"]


generate_coocurrance_analysis_plot(guardian_worker_rights_data.head(30),guardian_neg_words,guardian_pos_words,'Guardian','"Worker"')
generate_coocurrance_analysis_plot(bbc_worker_rights_data.head(30),bbc_neg_words,bbc_pos_words,'BBC','"Worker"')
generate_coocurrance_analysis_plot(aljazerra_worker_rights_data.head(30),aljazerra_neg_words,aljazerra_pos_words,'Al Jazeera', '"Worker"')
```

The findings reveal that in articles published by The Guardian, the most frequent co-occurring words with the word ”worker”
were negative words such as ”die”, ”kafala”, ”abuse”, and ”death”. In contrast, in articles published by Al Jazeera, the most common words co-occurring
with the word ”worker” were words related to legal reforms, such as ”change”, ”improve”,”ensure”, ”protect”, and ”protection”. Although the articles from Al Jazeera
also contained negative words such as ”death” and ”abuse”, they were not as
prevalent as in The Guardian articles.

As for the BBC, we also observed some negative words like ”death” appearing
alongside the word ”worker”, but to a lesser extent than in Al Jazeera articles.
Moreover, we noticed some positive words like ”reform”, ”ensure”, and ”change” in the articles, but not as frequently as in Al Jazeera articles. These findings could potentially provide an explanation for the reason why Al Jazeera tends to have a higher degree of negativity in its headlines compared to the BBC, but a lower level of negativity compared to The Guardian. Additionally, these results suggest that Al Jazeera has a more positive tone in its headlines compared to both of these media outlets.
