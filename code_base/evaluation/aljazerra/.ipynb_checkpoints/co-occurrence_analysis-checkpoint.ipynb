```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
aljazerra_data  = pd.read_csv("aljazerra_worker_rights.csv")
bbc_data = pd.read_csv("bbc_worker_rights.csv")
guardian_data = pd.read_csv("guardian_worker_rights.csv")
```

```python
def generate_coocurrance_analysis_plot(data,negative_words,positive_words):
    
    fig, ax = plt.subplots(figsize=(10, 9))

    colors = ['green' if x in positive_words else 'red' if x in negative_words else 'gray' for x in data['names']]

    # Plot the DataFrame
    plt.barh(data['names'], data['values'], color=colors )

    # Set the axis labels and title
    ax.set_xlabel('Frequency')
    ax.set_ylabel('Label')
    ax.set_title('co occurrence Analysis of the word "worker"')

    plt.show()
```

```python
guardian_neg_words = ["abuse","corruption","discriminatory","treatment","die","guardian","kafala","amnesty",
                       "death","exploit","exploitation"]
guardian_pos_words = ["protect","change","reform","improve"]

bbc_neg_words = ["wage","compensation","amnesty","death","abuse"]
bbc_pos_words= ["reform","ensure","change"]

aljazerra_pos_words = ["change","improve","ensure","protect","improve","protection"]
aljazerra_neg_words = ["abuse","complaint","death","salary"]


generate_coocurrance_analysis_plot(guardian_data.head(30),guardian_neg_words,guardian_pos_words)
generate_coocurrance_analysis_plot(bbc_data.head(30),bbc_neg_words,bbc_pos_words)
generate_coocurrance_analysis_plot(aljazerra_data.head(30),aljazerra_neg_words,aljazerra_pos_words)
```
